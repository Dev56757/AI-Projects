```python
# Write a while loop that terminates when a user types "exit".
user_input = ""
while (user_input.lower() != "exit"):
    user_input = input("Enter something (type 'exit' to quit): ")
print("Program terminated.")
```

```python
# Create a dictionary of 5 students and their marks; print only those who passed (>35).

students = {
    "Rahul": 78,
    "Anu": 25,
    "Kiran": 90,
    "Sneha": 40,
    "Ravi": 32
}
print("Students who passed:")
for name, marks in students.items():
    if marks > 35:
        print(name, ":", marks)
```

```python
# Use a list comprehension to square all numbers in a list.

# Original list
numbers = [1, 2, 3, 4, 5]
# List comprehension to square each number
squares = [num ** 2 for num in numbers]
print("Squared numbers:", squares)
```

```python
# How do you merge two dictionaries in Python 3.9+?

dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
# Merge using the | operator (Python 3.9+)
merged_dict = dict1 | dict2
print("Merged Dictionary:", merged_dict)
```

```python
# Write a script to find the frequency of each character in a string.

text = input("Enter a string: ")
frequency = {}
for char in text:
    frequency[char] = frequency.get(char, 0) + 1
print("Character Frequencies:")
for char, count in frequency.items():
    print(char, ":", count)
```

```python
# Demonstrate the enumerate() function in a for loop.

Animals = ["Lion", "Tiger", "Elephant", "Wolf"]
for index, Animals in enumerate(Animals):
    print("Index:", index, "Animal:", Animals)
```

```python
# Write a program to find the intersection of two sets.

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
intersection_set = set1.intersection(set2)
print("Intersection:", intersection_set)
```

```python
# Use zip() to combine two lists into a dictionary.

keys = ["id", "name", "age"]
values = [100, "Rishi", 21]
student_dict = dict(zip(keys, values))
print("Dictionary:", student_dict)
```

```python
# Create a 3x3 matrix using nested lists and access the middle element.
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
middle_element = matrix[1][1]
print("Matrix:")
for row in matrix:
    print(row)
print("Middle Element:", middle_element)
```

```python
# Write a script to sort a dictionary by its values.
marks = {
    "Anushka": 96,
    "Vikram": 55,
    "Shruthi": 99,
    "Jahnavi": 35,
    "Rishi" : 98
}
sorted_marks = dict(sorted(marks.items(), key=lambda item: item[1]))
print("Dictionary sorted by values:")
for name, mark in sorted_marks.items():
    print(name, ":", mark)
```
