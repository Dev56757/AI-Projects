```python
# Write a code snippet to remove duplicates from a list using a set.
# Code
numbers = [10, 20, 30, 20, 40, 10, 50]
unique_numbers = list(set(numbers))
print("List without duplicates:", unique_numbers)
```

```python
# How do you check if a specific key exists in a dictionary?
# The 'in' keyword is used to check whether a key exists in a dictionary.
# Code:
student = {
    "name": "Rahul",
    "age": 20,
    "course": "Python"
}
if "age" in student:
    print("Key exists")
else:
    print("Key does not exist")
```

```python
# Explain the difference between is and ==.
# '==' checks whether two values are equal & 'is' checks whether two variables refer to the same object in memory.
# Example Code:
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(a == b)
print(a is b)
print(a is c)
```

```python
# Create a simple calculator using if-elif-else logic.
# Code:
n1 = float(input("First number: "))
n2 = float(input("Second number: "))
op = input("Enter operator (+, -, *, /): ")
if op == "+":
    print("Result =", n1 + n2)
elif op == "-":
    print("Result =", n1 - n2)
elif op == "*":
    print("Result =", n1 * n2)
elif op == "/":
    if n2 != 0:
        print("Result =", n1 / n2)
    else:
        print("Division by zero is not allowed")
else:
    print("Invalid operator")
```

```python
# Write a program to find the largest number in a list without using max().
# Code:
numbers = [12, 45, 7, 89, 34]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num
print("Largest number is:", largest)
```

```python
# Demonstrate how to use the .strip() method on a messy input string.
# Code
text = " Hello Python "
clean_text = text.strip()
print("Before:", text)
print("After :", clean_text)
```

```python
# How do you convert a float to an integer? What happens to the decimal?
# The int() function converts a float into an integer by removing the decimal part.
# Code:
num = 25.89
integer_num = int(num)
print("Float value:", num)
print("Integer value:", integer_num)
```

```python
# Write a script that prints the multiplication table of a given number.
# Code:
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# Explain the use of the in keyword in Python.
# The 'in' keyword checks whether a value exists in a sequence such as a list, tuple, string, or dictionary.
# Code:
fruits = ["apple", "banana", "mango"]
if "banana" in fruits:
    print("Banana is present")
else:
    print("Banana is not present")
```
