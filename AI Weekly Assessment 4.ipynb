```python
# Explain the break, continue, and pass statements with examples.
# break: Terminates the loop immediately.
# continue: Skips the present iteration and moves to next iteration.
# pass: Acts as placeholder.
# Example Code:
print("Example of break:")
for i in range(1, 6):
    if i == 3:
        break
    print(i)
print("Example of continue:")
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
print("Example of pass:")
for i in range(1, 6):
    if i == 3:
        pass
    print(i)
```

```python
# Convert a list of temperatures in Celsius to Fahrenheit using map().
# Code:

# List of temperatures
celsius = [0, 20, 30, 40, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Celsius Temperatures :", celsius)
print("Fahrenheit Temperatures :", fahrenheit)
```

```python
# Find the second smallest element in a list.
# Code:

# Input list
numbers = [12, 45, 2, 9, 7, 3]
unique_numbers = sorted(set(numbers))

# Find the second smallest element
second_smallest = unique_numbers[1]

print("Original List :", numbers)
print("Second Smallest Element :", second_smallest)
```

```python
# Write a program to check if a year is a leap year.
# Code:
year = int(input("Enter a year: "))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
```

```python
# How do you reverse a dictionary (keys become values and vice versa)?
# Code:

# Original dictionary
student = {
    "Name": "Rahul",
    "Age": 20,
    "Course": "B.Tech"
}

# Reversed dictionary
reversed_dict = {value: key for key, value in student.items()}

print("Original Dictionary :", student)
print("Reversed Dictionary :", reversed_dict)
```

```python
# Demonstrate the use of any() and all() on a list of booleans.
# Code:

# boolean values
values = [True, False, True, True]

# checking any() values true
print("Result of any() :", any(values))

# checking all() values true
print("Result of all() :", all(values))
```

```python
# Write a script to flatten a nested list (e.g., [[1,2],[3,4]] to [1,2,3,4]).
# Code:
nested_list = [[1, 2], [3, 4]]
flat_list = [item for sublist in nested_list for item in sublist]
print("Nested List :", nested_list)
print("Flattened List :", flat_list)
```

```python
# Explain the difference between append() and extend() in lists.
# append(): Adds 1 element at end.
# extend(): Adds all elements of another iterable and each element is added individually.
# Code:
list1 = [1, 2, 3]
list1.append([4, 5])
print("After append() :", list1)
list2 = [1, 2, 3]
list2.extend([4, 5])
print("After extend() :", list2)
```

```python
# Write a script to find the "Longest Word" in a sentence.
# Code:
sentence = input("Enter a sentence: ")
words = sentence.split()
longest_word = max(words, key=len)
print("Longest Word :", longest_word)
```

```python
# Use the 'collections.Counter' module to find the 3 most common items in a list.
# Code:
from collections import Counter

# Sample list
items = ['berry', 'banana', 'berry', 'jackfruit', 'mango', 'berry', 'papaya', 'banana', 'mango']

counter = Counter(items)
print("Three Most Common Items:")
print(counter.most_common(3))
```
