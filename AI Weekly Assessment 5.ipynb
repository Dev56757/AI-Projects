```python
# Write a function that accepts variable-length arguments (*args).
# Code:
def display_numbers(*args):
  print("Arguments received:", args)
  print("Sum =", sum(args))
display_numbers(10, 20, 30, 40, 50)
```

```python
#  Create a function that accepts keyword arguments (kwargs) and prints them.
# Code:
def student_details(**kwargs):
  for key, value in kwargs.items():
    print(f"{key} : {value}")
student_details(Name="Rishi", Age=20, Dept="CSE", CGPA=8.3)
```

```python
#  Write a recursive function to calculate the factorial of a number.
# Code:
def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n - 1)
print("Factorial =", factorial(4))
```

```python
# Write a lambda function to sort a list of tuples based on the second element.
# Code:
students = [("A", 80), ("B", 65), ("C", 95), ("D", 72)]
sorted_list = sorted(students, key=lambda x: x[1])
print("Sorted List:")
print(sorted_list)
```

```python
# Create a "Filter" function to extract prime numbers from a list.
# Code:
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
      if num % i == 0:
        return False
    return True

numbers = [2, 4, 5, 6, 7, 9, 11, 15, 17]
prime_numbers = list(filter(is_prime, numbers))
print("Prime Numbers:", prime_numbers)
```

```python
# What is a "Docstring" and how do you access it at runtime?
# Docstring: A docstring is defined as a string, that is used to describe a function, class or module. And it is accessed using '__doc__' attribute.
# Example Code:
def add(a, b):
  """This function adds two numbers."""
  return a + b
print(add.__doc__)
```

```python
# Demonstrate the use of nonlocal in a nested function.
# Code:
def outer_function():
  x = 100
  def inner_function():
    nonlocal x
    x = 200
    print("Inner Function:", x)
  inner_function()
  print("Outer Function:", x)
outer_function()
```

```python
#  Create a function decorator that prints "Starting..." before any function runs
# Code:
def start_decorator(func):
  def wrapper():
    print("Starting...")
    func()
  return wrapper
@start_decorator
def my_function():
  print("Lost in lostworld!")
my_function()
```
