```python
# Difference between list and tuple:
# List is mutable, that means its elements can be changed after creation.
# Tuple is immutable, that means its elements cannot be changed after creation.
# Memory Usage Example: Tuples consume less memory, because they are immutable.
# Example Code:
import sys
my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)
print("List Memory:", sys.getsizeof(my_list), "bytes")
print("Tuple Memory:", sys.getsizeof(my_tuple), "bytes")
```

```python
# Swapping of two variables in a single line, without third variable:
# Python supports multiple assignment, allowing variables to be swapped directly.
# Example code:
a = 10
b = 20
a, b = b, a
print("a =", a)
print("b =", b)
```

```python
# Data Type identification of each element in a mixed list
mixed_list = [18, 6.05, "AI", True, [1, 2]]
for item in mixed_list:
    print(item, "->", type(item))
```

```python
# Palindrome
text = input("Enter a string: ")
if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
# Difference between / and // operators:
# / - Returns floating-point division, // - Returns floor (integer) division.
# Example Code:
a = 10
b = 3
print("Using / :", a / b)
print("Using //:", a // b)
```

```python
# Extraction of Last 3 Characters of a String Using Slicing:
# It is done where, negative indexing starts from the end of the string.
# Example code:
text = "Programming"
last_three = text[-3:]
print("Last 3 characters:", last_three)
```

```python
# Count number of vowels in a User-Provided Sentence
sentence = input("Enter a sentence: ")
vowels = 'aeiouAEIOU'
count = 0
for char in sentence:
    if char in vowels:
        count += 1
print("Number of vowels:", count)
```

```python
# f-Strings:
# f-strings (formatted strings) allow variables and expressions to be inserted directly into strings using {}.
# Advantages include 'easy to read', 'direct evaluation of expressions'.
# Example Code:
name = "Rushyendra"
age = 20
print(f"My name is {name} and I am {age} years old.")
```

```python
# List creation and printing its even's:
numbers = list(range(1, 21))
for num in numbers:
  if(num%2==0):
    print(num)
```

```python
# Difference between range(5) and list(range(5)):
# range(5) - Creates a range object.
# list(range(5)) - Converts range object into a list.
# Example Code:
print(range(5))
print(list(range(5)))
```
