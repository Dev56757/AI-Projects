```python
#Performing EDA on Spotify Songs Dataset
## Dataset Source
## https://www.kaggle.com/datasets/maharshipandya/-spotify-tracks-dataset
```

```python
# Importing necessary libraries
# Pandas is used for data loading, cleaning, and manipulation.
# NumPy is used for numerical computations and array operations.
# Matplotlib is used to create basic charts and plots.
# Seaborn is used to create advanced statistical visualizations and improve plot aesthetics.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset Loading
# Loading the data is the first step in Exploratory Data Analysis (EDA) because it allows us to inspect the dataset structure, identify potential issues, and understand the available features before performing any cleaning or analysis.
df = pd.read_csv("/content/Spotify_Dataset.csv")
df
```

```python
# Displaying Initial Records
#The first few rows of the dataset are displayed using the .head() function to obtain a quick overview of the data.
df.head()
```

```python
# Inspecting Dataset Information
# The `df.info()` function provides a concise summary of the dataset, including column names, data types, and non-null value counts.
df.info()
```

```python
# Generating Statistical Summary
## The `df.describe()` function provides descriptive statistics for numerical columns in the dataset.

df.describe()
```

```python
# Checking Dataset Dimensions
## The `df.shape` function is used to determine the number of rows and columns present in the dataset.
df.shape
```

```python
# Viewing Column Names
## The `df.columns` function is used to display all feature names in the dataset.
df.columns
```

```python
# Missing Value Analysis
# Check missing values in each column
df.isnull().sum()
```

```python
# Fill numerical missing values with median
df['popularity'] = df['popularity'].fillna(df['popularity'].median())
print(df.isnull().sum())


```

```python
#Duplicate Value Analysis
# Count duplicate rows
print(df.duplicated().sum())
```

```python
# Data Type Correction
# Check current data types
df.dtypes
```

```python
# Data Type Correction
# The data types of all columns were examined to ensure that each feature is stored in an appropriate format.
# The dataset contains numerical, categorical, and boolean features. No date-related columns were present in the dataset; therefore, datetime conversion was not required and not possible.
# The categorical columns such as artists, album_name, track_name, and track_genre are stored as object data types, which are appropriate for textual data and can be converted to category type if memory optimization is needed.
```

```python
# Univariate Analysis
# Univariate analysis focuses on examining one variable at a time to understand its distribution, central tendency, variability, and frequency patterns.
# Numerical Features
# Outlier detection using sns.histplot()
## Popularity Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['popularity'], bins=30, kde=True)
plt.title('Distribution of Popularity')
plt.show()
```

```python
## Danceability Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['danceability'], bins=20, kde=True)
plt.title('Distribution of danceability')
plt.show()
```

```python
## Energy Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['energy'], bins=40, kde=True)
plt.title('Distribution of energy')
plt.show()
```

```python
## Tempo Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['tempo'], bins=40, kde=True)
plt.title('Distribution of tempo')
plt.show()

```

```python
# Categorical Features
## Genre Distribution
### Genre is an important categorical feature. Analyzing genre frequencies helps identify the most common and least common music genres in the dataset.
df['track_genre'].value_counts()
```

```python
# Count Plot for Top Genres
top_genres = df['track_genre'].value_counts().head(10)
plt.figure(figsize=(11,7))
sns.countplot(
    y='track_genre',
    data=df,
    order=df['track_genre'].value_counts().head(10).index
)
plt.title('Top 10 Music Genres')
plt.show()
```

```python
# Count Plot for Popularity Frequency Analysis
plt.figure(figsize=(15,12))
sns.countplot(x='popularity', data = df)
plt.title('Popularity Frequency Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bivariate & Multivariate Analysis
## Correlation Matrix
### Using df.corr() associated with sns.heatmap to detect dependencies
corr = df.corr(numeric_only = True)
plt.figure(figsize =(15,12))
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.show()
```

```python
# Scatter Plot 1
## To visualize relationships between two numeric variables
plt.figure(figsize=(15,12))
sns.scatterplot(x='energy', y='popularity', data=df)
plt.title('Energy vs Popularity')
plt.show()
```

```python
# Scatter Plot 2
plt.figure(figsize=(15,12))
sns.scatterplot(x='danceability', y='popularity', data=df)
plt.title('Danceability vs Popularity')
plt.show()
```

```python
# Box Plot
## Danceability across Genres
top_genres = df['track_genre'].value_counts().head(10).index
plt.figure(figsize=(14,6))
sns.boxplot(
    x='track_genre',
    y='danceability',
    data=df[df['track_genre'].isin(top_genres)]
)
plt.xticks(rotation=45)
plt.title('Danceability Across Top 10 Genres')
plt.show()
```

```python
#pairplot
## A pair plot visualizes relationships among multiple numerical variables simultaneously.
sns.pairplot(
    df[['popularity',
        'danceability',
        'energy',
        'tempo',
        'valence']]
)
plt.show()
```
