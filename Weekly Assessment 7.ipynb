```python
# Define a class Car with attributes and a method to display details.
# Code:
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    def display_details(self):
        print("Car Brand :", self.brand)
        print("Car Model :", self.model)
        print("Car Year  :", self.year)
car1 = Car("Toyota", "Innova", 2010)
car1.display_details()
```

```python
# Explain the purpose of the __init__ method.
# The __init__() method is called the constructor in Python. It initializes the object's attributes.
# Example Code:
class Car:
  def __init__(self, brand, model, year):
    self.brand = brand
    self.model = model
    self.year = year
car1 = Car("Toyota", "Innova", 2010)
```

```python
# Create a subclass ElectricCar that inherits from Car.
# Code:
class Car:
  def __init__(self, brand, model, year):
    self.brand = brand
    self.model = model
    self.year = year
  def display_details(self):
        print("Car Brand :", self.brand)
        print("Car Model :", self.model)
        print("Car Year  :", self.year)
class ElectricCar(Car):
  def __init__(self, brand, model, year, battery_capacity):
    super().__init__(brand, model, year)
    self.battery_capacity = battery_capacity
  def display_details(self):
    super().display_details()
    print("Battery Capacity:", self.battery_capacity)
Car1 = Car("Hyundai", "Creta", "2018")
Car1.display_details()
Car2 = ElectricCar("Tesla", "Model S", "2020", "100 kWh")
Car2.display_details()
```

```python
# Demonstrate "Method Overriding" in inheritance.
# Code:
# Parent Class
class Vehicle:
  def pollution(self):
    print("Vehicle produces pollution.")
# Child Class
class Lorry(Vehicle):
  # Method Overriding
  def pollution(self):
    print("Lorry produces pollution and high horn sounds.")
L = Lorry()
L.pollution()
```

```python
# Explain "Encapsulation" by creating private attributes using __.
# Encapsulation is the process of hiding data and allowing access only through methods.
# Code:
class Student:
  def __init__(self, name, marks):
    self.name = name
    # Private data
    self.__marks = marks
  def display(self):
    print("Marks :", self.__marks)
s = Student("Shruthi", 94)
s.display()
```

```python
# Use @property to create getter and setter methods.
# Code:
class Student:
  def __init__(self, name, marks):
    self.name = name
    self.__marks = marks
  @property
  def marks(self):
    return self.__marks
  @marks.setter
  def marks(self, marks):
    self.__marks = marks
s = Student("Maya", 66)
print("Marks :", s.marks)
s.marks = 86
print("Marks :", s.marks)
```

```python
# What is "Multiple Inheritance"? Provide a small code structure.
# Multiple inheritance means a child class inherits properties and methods from more than one parent class.
# Code:
# Parent Class 1
class A:
  def A_method(self):
    print("This comprises method of class A.")
# Parent Class 2
class B:
  def B_method(self):
    print("This comprises method of class B.")
# Child Class
class C(A, B):
  def C_method(self):
    print("This comprises method of class C.")
c = C()
c.A_method()
c.B_method()
c.C_method()
```

```python
# Create a class method using the @classmethod decorator.
# Code:
class School:
  school_name = "XYZ School"
  @classmethod
  def display_school(cls):
    print("School Name :", cls.school_name)
School.display_school()
```

```python
# Explain the difference between a class method and a static method.
# Class method: Uses '@classmethod' decorator.
# Static method: Uses '@staticmethod' decorator.
# Code:
class School:
  school_name = "XYZ School"
  @classmethod
  def display_school(cls):
    print("School Name :", cls.school_name)
  @staticmethod
  def display_details():
    print("This is static method.")
School.display_school()
School.display_details()
```
